### 니트웨어

```python
##남성 니트웨어류 크롤링 
man_knitwear = 'https://www.mytheresa.com/kr/ko/men/clothing/knitwear'
driver = webdriver.Chrome()
driver.get(man_knitwear)
```

```python
# 데이터를 저장할 리스트
knit_list = []

# 현재 페이지 번호
current_page = 1

try:
    while True:
        # 현재 페이지의 이미지 정보 수집
        for i in range(1, 61):
            xpath = f'//*[@id="root"]/div/div[2]/div[1]/div[4]/div[{i}]/a/div[1]/div[1]/img'
            try:
                # 이미지가 나타날 때까지 기다림 (최대 10초)
                WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, xpath)))
                image_element = driver.find_element(By.XPATH, xpath)
                man_knit_name = image_element.get_attribute('alt')
                man_knit_img = image_element.get_attribute('src')
                
                # 데이터 리스트에 추가
                knit_list.append({'Name': man_knit_name, 'Image URL': man_knit_img})
            
            except TimeoutException:
                print(f'Timed out waiting for element at xpath: {xpath}')
                continue  # 요소를 찾지 못하면 다음 반복으로 넘어감
            except NoSuchElementException:
                print(f'Element not found at xpath: {xpath}. Skipping.')
                continue  # 요소를 찾지 못하면 다음 반복으로 넘어감
        
        # 다음 페이지로 이동
        next_button_xpath = '//*[@id="root"]/div/div[2]/div[1]/div[5]/div/div[3]/a[1]'
        
        # 다음 페이지 버튼이 있는지 확인
        next_button = driver.find_elements(By.XPATH, next_button_xpath)
        if len(next_button) == 0:
            print(f'No more next button found. Stopping scraping.')
            break
        
        # 버튼이 클릭 가능할 때까지 대기하고 클릭
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, next_button_xpath)))
        
        # 다음 버튼이 가려져 있을 경우 스크롤하여 가시성 조정
        driver.execute_script("arguments[0].scrollIntoView();", next_button[0])
        time.sleep(1)  # 스크롤 후 대기
        
        # 클릭 시도
        try:
            next_button[0].click()
        except ElementClickInterceptedException:
            print(f'ElementClickInterceptedException: Click intercepted. Trying to click with JS.')
            driver.execute_script("arguments[0].click();", next_button[0])
        
        # 페이지 번호 업데이트
        current_page += 1
        
        # 페이지 로드 대기 (필요시 조정)
        time.sleep(2)

except NoSuchElementException:
    print(f'No more pages available. Stopping scraping.')

finally:
    # 브라우저 종료
    driver.quit()

    # 데이터를 pandas DataFrame으로 변환
    knit = pd.DataFrame(knit_list)
```

```python
# 엑셀 파일로 저장
knit.to_excel('man_knitwear.xlsx', index=False)
```

```python
# 데이터를 저장할 리스트
dress_list = []

# 현재 페이지 번호
current_page = 1

try:
    while True:
        # 현재 페이지의 이미지 정보 수집
        for i in range(1, 61):
            xpath = f'//*[@id="root"]/div/div[2]/div[1]/div[4]/div[{i}]/a/div[1]/div[1]/img'
            try:
                # 이미지가 나타날 때까지 기다림 (최대 10초)
                WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, xpath)))
                image_element = driver.find_element(By.XPATH, xpath)
                women_dress_name = image_element.get_attribute('alt')
                women_dress_img = image_element.get_attribute('src')
                
                # 데이터 리스트에 추가
                dress_list.append({'Name': women_dress_name, 'Image URL': women_dress_img})
            
            except TimeoutException:
                print(f'Timed out waiting for element at xpath: {xpath}')
                continue  # 요소를 찾지 못하면 다음 반복으로 넘어감
            except NoSuchElementException:
                print(f'Element not found at xpath: {xpath}. Skipping.')
                continue  # 요소를 찾지 못하면 다음 반복으로 넘어감
        
        # 다음 페이지로 이동
        next_button_xpath = '//*[@id="root"]/div/div[2]/div[1]/div[5]/div/div[3]/a[1]'
        
        # 다음 페이지 버튼이 있는지 확인
        next_button = driver.find_elements(By.XPATH, next_button_xpath)
        if len(next_button) == 0:
            print(f'No more next button found. Stopping scraping.')
            break
        
        # 버튼이 클릭 가능할 때까지 대기하고 클릭
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, next_button_xpath)))
        
        # 다음 버튼이 가려져 있을 경우 스크롤하여 가시성 조정
        driver.execute_script("arguments[0].scrollIntoView();", next_button[0])
        time.sleep(1)  # 스크롤 후 대기
        
        # 클릭 시도
        try:
            next_button[0].click()
        except ElementClickInterceptedException:
            print(f'ElementClickInterceptedException: Click intercepted. Trying to click with JS.')
            driver.execute_script("arguments[0].click();", next_button[0])
        
        # 페이지 번호 업데이트
        current_page += 1
        
        # 페이지 로드 대기 (필요시 조정)
        time.sleep(2)

except NoSuchElementException:
    print(f'No more pages available. Stopping scraping.')

finally:
    # 브라우저 종료
    driver.quit()

    # 데이터를 pandas DataFrame으로 변환
    dress = pd.DataFrame(dress_list)
```

```python
##여성 투피스 세트 크롤링 
women_twopieceset = 'https://www.mytheresa.com/kr/ko/women/clothing/two-piece-sets'
driver = webdriver.Chrome()
driver.get(women_twopieceset)
```

```python
# 데이터를 저장할 리스트
twopieceset_list = []

# 현재 페이지 번호
current_page = 1

try:
    while True:
        # 현재 페이지의 이미지 정보 수집
        for i in range(1, 61):
            xpath = f'//*[@id="root"]/div/div[2]/div[1]/div[4]/div[{i}]/a/div[1]/div[1]/img'
            try:
                # 이미지가 나타날 때까지 기다림 (최대 10초)
                WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, xpath)))
                image_element = driver.find_element(By.XPATH, xpath)
                women_twopieceset_name = image_element.get_attribute('alt')
                women_twopieceset_img = image_element.get_attribute('src')
                
                # 데이터 리스트에 추가
                twopieceset_list.append({'Name': women_twopieceset_name, 'Image URL': women_twopieceset_img})
            
            except TimeoutException:
                print(f'Timed out waiting for element at xpath: {xpath}')
                continue  # 요소를 찾지 못하면 다음 반복으로 넘어감
            except NoSuchElementException:
                print(f'Element not found at xpath: {xpath}. Skipping.')
                continue  # 요소를 찾지 못하면 다음 반복으로 넘어감
        
        # 다음 페이지로 이동
        next_button_xpath = '//*[@id="root"]/div/div[2]/div[1]/div[5]/div/div[3]/a[1]'
        
        # 다음 페이지 버튼이 있는지 확인
        next_button = driver.find_elements(By.XPATH, next_button_xpath)
        if len(next_button) == 0:
            print(f'No more next button found. Stopping scraping.')
            break
        
        # 버튼이 클릭 가능할 때까지 대기하고 클릭
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, next_button_xpath)))
        
        # 다음 버튼이 가려져 있을 경우 스크롤하여 가시성 조정
        driver.execute_script("arguments[0].scrollIntoView();", next_button[0])
        time.sleep(1)  # 스크롤 후 대기
        
        # 클릭 시도
        try:
            next_button[0].click()
        except ElementClickInterceptedException:
            print(f'ElementClickInterceptedException: Click intercepted. Trying to click with JS.')
            driver.execute_script("arguments[0].click();", next_button[0])
        
        # 페이지 번호 업데이트
        current_page += 1
        
        # 페이지 로드 대기 (필요시 조정)
        time.sleep(2)

except NoSuchElementException:
    print(f'No more pages available. Stopping scraping.')

finally:
    # 브라우저 종료
    driver.quit()

    # 데이터를 pandas DataFrame으로 변환
    twopieceset = pd.DataFrame(twopieceset_list)
```
